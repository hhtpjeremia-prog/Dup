```python
import kagglehub
import os
import pandas as pd
import glob
import warnings
warnings.filterwarnings('ignore')
```

```python
DatasetPath = kagglehub.dataset_download("geraldooizx/g-coffee-shop-transaction-202307-to-202506")
```

```python
def load_multiple_csvs(pattern):
    files = glob.glob(pattern)
    return pd.concat((pd.read_csv(f) for f in files), ignore_index=True)

df_Trans = load_multiple_csvs(os.path.join(DatasetPath, "transactions", "*.csv"))
df_TransItem = load_multiple_csvs(os.path.join(DatasetPath, "transaction_items", "*.csv"))
df_Users = load_multiple_csvs(os.path.join(DatasetPath, "users", "*.csv"))

df_menu = pd.read_csv(os.path.join(DatasetPath, "menu_items", "menu_items.csv"))
df_stores = pd.read_csv(os.path.join(DatasetPath, "stores", "stores.csv"))
df_vouchers = pd.read_csv(os.path.join(DatasetPath, "vouchers", "vouchers.csv"))
df_payments = pd.read_csv(os.path.join(DatasetPath, "payment_methods", "payment_methods.csv"))
```
